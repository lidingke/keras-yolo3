```python
import mysql.connector
from sqlalchemy import create_engine
import os
engine = create_engine('mysql+mysqlconnector://trnet:trnet@127.0.0.1:3306/trnet')
con = engine.connect()
```

```python
def sort_char(sq_cs):
    collections = {}
    for s in sq_cs:
        if s not in collections.keys():
            collections[s] = 1
        else:
            collections[s] = 1 + collections[s]
    return list(collections.items()).copy()

def print_result(merge_data):
#     print('{}栏位：'.format(label))
    merge_data_sorted = merge_data.sort(key=lambda x :x[1],reverse=True)
    merge_data_sum = sum([int(d[1]) for d in merge_data])
    merge_data_sum_100 = len([int(d[1]) for d in merge_data if int(d[1]) > 1])
    print("汉字分类数：{}，汉字总数：{},其中超过100的字符个数{}".format(len(merge_data),merge_data_sum,merge_data_sum_100))
    for k,v in merge_data[:50]:
        print(u"{}:{}".format(k,v))
def get_3755char():
    with open('./3755.txt', 'rb') as f:
        c = f.read()
    cd = c.decode('utf-8')
    return list(cd)
```

```python
%%time
with engine.connect() as con:
#     d = con.execute("SELECT content FROM data04 WHERE id=48519;")
#     d = con.execute("SELECT id FROM data04;")
    cmd = "SELECT content FROM data04 WHERE label in ('FD0007','FD0032','FD0033','FD0076','FD0147');"
    hccr_lst04 = con.execute(cmd)
    cmd = "SELECT content FROM data05 WHERE label in ('FD0007','FD0032','FD0033','FD0076','FD0147');"
    hccr_lst05 = con.execute(cmd)
chars_04 = "".join(h[0] for h in hccr_lst04)
chars_05 = "".join(h[0] for h in hccr_lst05)
chars = chars_04+chars_05
char_3755 = get_3755char()
data = sort_char(chars)
data3755 = [d for d in data if d[0] in char_3755]
dataout_3755 = [d for d in data if (d[0] not in
                                    char_3755) and ord(d[0])>255]
# print_result(data)
print('3755之内：')
print_result(data3755)
print('3755之外：')
print_result(dataout_3755)
```
